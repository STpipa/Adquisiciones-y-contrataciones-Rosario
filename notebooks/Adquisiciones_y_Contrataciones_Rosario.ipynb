## Limpieza Adicional de Datos

Basándonos en la información obtenida de las celdas anteriores, es necesario limpiar las columnas que representan valores monetarios ('Monto Oficial', 'Costo Pliego', 'Sellado') para convertirlas a un formato numérico. También eliminaremos las columnas que tienen un alto porcentaje de valores faltantes y que no son relevantes para los análisis principales.

```python
# Función para limpiar y convertir columnas monetarias
def limpiar_monto(df, columna):
    if columna in df.columns:
        # Rellenar valores faltantes con '0,00' antes de limpiar
        df[columna] = df[columna].fillna('0,00')
        # Eliminar puntos como separador de miles y reemplazar coma por punto como separador decimal
        df[columna] = df[columna].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        # Convertir a tipo numérico, forzando errores a NaN
        df[columna] = pd.to_numeric(df[columna], errors='coerce')
        # Rellenar los valores que no pudieron ser convertidos (NaN) con 0
        df[columna] = df[columna].fillna(0)
    else:
        print(f"Advertencia: La columna '{columna}' no se encuentra en el DataFrame.")
    return df

# Limpiar columnas monetarias en ambos DataFrames
df_gasto_2023 = limpiar_monto(df_gasto_2023, 'Monto Oficial')
df_gasto_2023 = limpiar_monto(df_gasto_2023, 'Costo Pliego')
df_gasto_2023 = limpiar_monto(df_gasto_2023, 'Sellado')

df_gasto_2024 = limpiar_monto(df_gasto_2024, 'Monto Oficial')
df_gasto_2024 = limpiar_monto(df_gasto_2024, 'Costo Pliego')
df_gasto_2024 = limpiar_monto(df_gasto_2024, 'Sellado')

print("Columnas monetarias limpiadas y convertidas a numérico.")
```

Ahora, eliminamos las columnas con una gran cantidad de valores faltantes que no aportan al análisis principal.

```python
columnas_a_eliminar = ['Observaciones', 'Tipo de Norma', 'Nro de Norma', 'Año Norma', 'Unnamed: 18']

# Eliminar columnas en ambos DataFrames si existen
columnas_existentes_2023 = [col for col in columnas_a_eliminar if col in df_gasto_2023.columns]
df_gasto_2023 = df_gasto_2023.drop(columns=columnas_existentes_2023)
if columnas_existentes_2023:
    print(f"Columnas eliminadas en DataFrame 2023: {columnas_existentes_2023}")
else:
    print("Ninguna de las columnas a eliminar se encontró en DataFrame 2023.")


columnas_existentes_2024 = [col for col in columnas_a_eliminar if col in df_gasto_2024.columns]
df_gasto_2024 = df_gasto_2024.drop(columns=columnas_existentes_2024)
if columnas_existentes_2024:
    print(f"Columnas eliminadas en DataFrame 2024: {columnas_existentes_2024}")
else:
    print("Ninguna de las columnas a eliminar se encontró en DataFrame 2024.")

print("\nColumnas irrelevantes con muchos valores faltantes eliminadas.")
```

## Análisis de Gasto por Área

Calcularemos el total y la cantidad de compras directas y erogaciones menores por cada 'Repartición Licitante' para los años 2023 y 2024.

Para este análisis, consideraremos los tipos de licitación que generalmente corresponden a compras directas o erogaciones menores. Basándonos en los datos mostrados anteriormente, los tipos relevantes podrían ser 'Contratación Directa' y 'Concurso de Precios'. Si existen otros tipos que deban incluirse, se pueden ajustar en el código.

```python
# Definir los tipos de licitación que consideramos como compras directas o menores
tipos_menores = ['Contratación Directa', 'Concurso de Precios']

# Filtrar los DataFrames para incluir solo los tipos de licitación menores
df_menores_2023 = df_gasto_2023[df_gasto_2023['Tipo de Licitación'].isin(tipos_menores)].copy()
df_menores_2024 = df_gasto_2024[df_gasto_2024['Tipo de Licitación'].isin(tipos_menores)].copy()

# Agrupar por 'Repartición Licitante' y calcular la cantidad de procesos y la suma del 'Monto Oficial'
gasto_por_area_2023 = df_menores_2023.groupby('Repartición Licitante').agg(
    cantidad_procesos=('Nro. Expediente', 'count'),
    monto_total=('Monto Oficial', 'sum')
).reset_index()

gasto_por_area_2024 = df_menores_2024.groupby('Repartición Licitante').agg(
    cantidad_procesos=('Nro. Expediente', 'count'),
    monto_total=('Monto Oficial', 'sum')
).reset_index()

print("Análisis de gasto por área para compras directas y menores (2023):")
display(gasto_por_area_2023.sort_values(by='monto_total', ascending=False).head(10)) # Mostrar las 10 áreas con mayor gasto

print("\nAnálisis de gasto por área para compras directas y menores (2024):")
display(gasto_por_area_2024.sort_values(by='monto_total', ascending=False).head(10)) # Mostrar las 10 áreas con mayor gasto
```

## Análisis de Modalidad de Gasto

Ahora, analizaremos la proporción del gasto anual total para cada tipo de licitación. Esto nos ayudará a entender cómo se distribuye el gasto público entre las diferentes modalidades de contratación (Licitación Pública, Licitación Privada, Concurso de Precios, Contratación Directa, etc.).

```python
# Calcular el gasto total por tipo de licitación para 2023
gasto_por_modalidad_2023 = df_gasto_2023.groupby('Tipo de Licitación')['Monto Oficial'].sum().reset_index()

# Calcular el porcentaje del gasto total para cada tipo de licitación en 2023
gasto_total_2023 = gasto_por_modalidad_2023['Monto Oficial'].sum()
gasto_por_modalidad_2023['Porcentaje'] = (gasto_por_modalidad_2023['Monto Oficial'] / gasto_total_2023) * 100

print("Análisis de gasto por modalidad de licitación (2023):")
display(gasto_por_modalidad_2023.sort_values(by='Monto Oficial', ascending=False))

# Calcular el gasto total por tipo de licitación para 2024
gasto_por_modalidad_2024 = df_gasto_2024.groupby('Tipo de Licitación')['Monto Oficial'].sum().reset_index()

# Calcular el porcentaje del gasto total para cada tipo de licitación en 2024
gasto_total_2024 = gasto_por_modalidad_2024['Monto Oficial'].sum()
gasto_por_modalidad_2024['Porcentaje'] = (gasto_por_modalidad_2024['Monto Oficial'] / gasto_total_2024) * 100

print("\nAnálisis de gasto por modalidad de licitación (2024):")
display(gasto_por_modalidad_2024.sort_values(by='Monto Oficial', ascending=False))
```

## Análisis de Repetición de Compras

Para analizar la repetición de compras, vamos a identificar si se realizaron adquisiciones del mismo 'Objeto' (o similar) en un corto período de tiempo.

Primero, prepararemos los datos combinando los años 2023 y 2024 en un solo DataFrame y asegurándonos de que las fechas estén en el formato correcto. Luego, implementaremos la lógica para identificar las compras repetidas.

```python
# Combinar los DataFrames de 2023 y 2024
df_gasto_combinado = pd.concat([df_gasto_2023, df_gasto_2024], ignore_index=True)

# Convertir las columnas de fecha a formato datetime
# Usamos errors='coerce' para convertir valores inválidos en NaT (Not a Time)
df_gasto_combinado['Fecha Apertura'] = pd.to_datetime(df_gasto_combinado['Fecha Apertura'], format='%d/%m/%Y', errors='coerce')
df_gasto_combinado['Fecha Resolución'] = pd.to_datetime(df_gasto_combinado['Fecha Resolución'], format='%d/%m/%Y', errors='coerce')

print("DataFrames de 2023 y 2024 combinados y columnas de fecha convertidas.")
```

Ahora, vamos a simplificar la columna 'Objeto' para agrupar objetos similares. Dada la naturaleza de los datos, una forma sencilla es convertir todo a minúsculas y eliminar espacios extra.

```python
# Limpiar la columna 'Objeto'
df_gasto_combinado['Objeto_limpio'] = df_gasto_combinado['Objeto'].astype(str).str.lower().str.strip()

print("Columna 'Objeto' limpiada.")
```

Ahora, identificamos las compras repetidas. Agrupamos por el 'Objeto_limpio' y ordenaremos por fecha de apertura. Luego, calculamos la diferencia de tiempo entre compras consecutivas del mismo objeto. Si la diferencia es menor a 90 días (o 30 para alimentos/medicamentos), marcaremos la compra como repetida.

```python
# Ordenar por objeto limpio y fecha de apertura
df_gasto_combinado_sorted = df_gasto_combinado.sort_values(by=['Objeto_limpio', 'Fecha Apertura'])

# Calcular la diferencia de tiempo entre compras consecutivas del mismo objeto
df_gasto_combinado_sorted['dias_entre_compras'] = df_gasto_combinado_sorted.groupby('Objeto_limpio')['Fecha Apertura'].diff().dt.days

# Identificar compras repetidas (diferencia menor a 90 días)
# Nota: Esta es una simplificación. Identificar alimentos/medicamentos requeriría análisis de texto más avanzado.
df_gasto_combinado_sorted['compra_repetida_90_dias'] = df_gasto_combinado_sorted['dias_entre_compras'] < 90

# Mostrar ejemplos de compras repetidas
print("Ejemplos de posibles compras repetidas (con menos de 90 días de diferencia):")
display(df_gasto_combinado_sorted[df_gasto_combinado_sorted['compra_repetida_90_dias']].head(10))

print("\nResumen de compras posiblemente repetidas (con menos de 90 días de diferencia):")
display(df_gasto_combinado_sorted['compra_repetida_90_dias'].value_counts())
```

## Análisis de Oferentes y Adjudicatarios

Vamos a analizar quiénes son los proveedores que ofertaron y a quiénes se les adjudicaron contratos. Extraeremos los nombres de los adjudicatarios de la columna 'Resultado' y luego analizaremos su participación y los montos adjudicados.

```python
# Función para intentar extraer el nombre del adjudicatario de la columna 'Resultado'
def extraer_adjudicatario(resultado):
    if isinstance(resultado, str):
        # Convertir a minúsculas para facilitar la búsqueda
        resultado_lower = resultado.lower()

        # Buscar patrones comunes que indiquen una adjudicación y el nombre de la empresa
        # Estos patrones pueden necesitar ajustarse según la variabilidad del texto
        patrones_adjudicado = [
            r'se adjudico a (.*?) por',
            r'se adjudicó a (.*?) por',
            r'adjudicase a la firma (.*?) por',
            r'adjudicase al adjudicatario: (.*?)\.', # para el patrón de la última fila de df_gasto_combinado_sorted
            r'a la firma:(.*?) por',
            r'adjudicado a:(.*?)',
            r'adjudicado a la firma:(.*?)',
            r'se adjudica a la firma (.*?) por', # Nuevo patrón basado en ejemplos
            r'se adjudica a (.*?)\.', # Nuevo patrón basado en ejemplos
            r'adjudicase\r\n(.*?)\r\n', # Nuevo patrón basado en ejemplos
            r'se adjudica a (.*?), cuit:', # Nuevo patrón basado en ejemplos
            r'adjudicase (.*?) item', # Nuevo patrón basado en ejemplos
            r'adjudicase por mejor precio a la firma de (.*?)\r\n', # Nuevo patrón basado en ejemplos
            r'por mejor precio a la firma (.*?)\r\n', # Nuevo patrón basado en ejemplos
            r'se adjudica a la firma (.*?) cuit', # Nuevo patrón basado en ejemplos

        ]

        for patron in patrones_adjudicado:
            match = re.search(patron, resultado_lower)
            if match:
                nombre_adjudicatario = match.group(1).strip()
                # Limpiar el nombre extraído (eliminar comas, puntos al final si existen)
                nombre_adjudicatario = nombre_adjudicatario.replace('.', '').replace(',', '')
                return nombre_adjudicatario.upper() # Convertir a mayúsculas para estandarizar
    return "NO ADJUDICADO / NO ESPECIFICADO" # Valor por defecto si no se encuentra adjudicatario

# Importar la librería de expresiones regulares
import re

# Aplicar la función para extraer adjudicatarios al DataFrame combinado
# Primero, rellenar los valores faltantes en 'Resultado' para evitar errores
df_gasto_combinado['Resultado'] = df_gasto_combinado['Resultado'].fillna('')
df_gasto_combinado['Adjudicatario'] = df_gasto_combinado['Resultado'].apply(extraer_adjudicatario)

# Mostrar los adjudicatarios más frecuentes para verificar la extracción
print("Adjudicatarios más frecuentes (ejemplos):")
display(df_gasto_combinado['Adjudicatario'].value_counts().head(10))
```

Ahora que hemos extraído (con la mejor aproximación posible dada la estructura de los datos) los nombres de los adjudicatarios, podemos analizar la cantidad de procesos y el monto total adjudicado a cada uno.

```python
# Filtrar para incluir solo los procesos que fueron adjudicados y donde se pudo extraer un adjudicatario
df_adjudicados = df_gasto_combinado[df_gasto_combinado['Adjudicatario'] != "NO ADJUDICADO / NO ESPECIFICADO"].copy()

# Agrupar por adjudicatario y calcular la cantidad de procesos adjudicados y la suma del monto oficial (como aproximación al monto adjudicado)
analisis_adjudicatarios = df_adjudicados.groupby('Adjudicatario').agg(
    cantidad_procesos_adjudicados=('Nro. Expediente', 'count'),
    monto_total_oficial_adjudicado=('Monto Oficial', 'sum') # Usamos Monto Oficial como proxy
).reset_index()

# Ordenar por monto total oficial adjudicado de forma descendente
analisis_adjudicatarios_sorted = analisis_adjudicatarios.sort_values(by='monto_total_oficial_adjudicado', ascending=False)

print("\nAnálisis de Adjudicatarios (Cantidad de Procesos y Monto Oficial Adjudicado):")
display(analisis_adjudicatarios_sorted.head(10)) # Mostrar los 10 adjudicatarios con mayor monto adjudicado
```

Este análisis nos da una primera vista de los principales proveedores en términos de cantidad de procesos y monto oficial (que usamos como proxy del monto adjudicado). Es importante recordar que la extracción de los nombres puede no ser perfecta debido a la variabilidad en el texto de la columna 'Resultado'.
